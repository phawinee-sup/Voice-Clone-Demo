```python
import os
import subprocess
import sys

def denoise_speech(input_path, output_dir):
    env = os.environ.copy()
    env["DF_NO_GIT"] = "1"

    result = subprocess.run(
        [
            sys.executable,
            "-m", "df.enhance",
            "-i", input_path,
            "-o", output_dir
        ],
        env=env,
        check=True
    )

```

```python
import os

folder_list = ["Lisa", "Bambam", "IU", "IVE"]
file_path_list = []

for folder in folder_list:
    sub_folder = folder + "/dataset"
    for file in os.listdir(sub_folder):
        folder_path = folder + "/dataset/" + file
        file_path_list.append(folder_path)

file_path_list[0]
```

```python
file_path_list[0].replace("dataset", "dataset_denoised")
```

```python
for file_path in file_path_list:
    input_path = file_path

    output_dir = file_path.replace("dataset", "dataset_denoised")

    os.makedirs(output_dir, exist_ok=True)

    denoise_speech(input_path, output_dir)

    print(f"Done: {input_path} -> {output_dir}")
```

```python
import json

with open("manifest_whisper.json", "r", encoding="utf-8") as f:
    manifest_whisper = json.load(f)

for data_dict in manifest_whisper:
    audio_path = data_dict["audio"]

    # ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô dataset -> dataset_denoised
    audio_path = audio_path.replace("dataset", "dataset_denoised")

    # ‡πÄ‡∏û‡∏¥‡πà‡∏° _DeepFilterNet3 ‡∏Å‡πà‡∏≠‡∏ô .wav
    audio_path = audio_path.replace(".wav", "_DeepFilterNet3.wav")

    data_dict["audio"] = audio_path

with open("manifest_whisper_rename_path.json", "w", encoding="utf-8") as f:
    json.dump(manifest_whisper, f, ensure_ascii=False, indent=2)
```
