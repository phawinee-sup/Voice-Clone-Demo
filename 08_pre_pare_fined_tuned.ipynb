```python
import os
import json
import pandas as pd
import ast
```

```python
df = pd.read_csv("transcription_phonemes.csv")
df.head()
```

```python
df[df['file_path'] == "Lisa/dataset/5QmaVImK37w/SPEAKER_06_sent_0000.wav"]["whisper_text"].iloc[0]
```

```python
with open("manifest_whisper.json", "r", encoding="utf-8") as f:
    manifest_whisper = json.load(f)

for data_dict in manifest_whisper:
    k = data_dict.pop("phonemes")
    file_path = data_dict["audio"]
    text = df[df['file_path'] == file_path]["whisper_text"].iloc[0]
    data_dict["text"] = text

with open("manifest_whisper_text.json", "w", encoding="utf-8") as f:
    json.dump(manifest_whisper, f, ensure_ascii=False, indent=2)
```
